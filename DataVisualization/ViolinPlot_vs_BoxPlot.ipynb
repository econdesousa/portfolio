We'll start by importing standard packages: 'pandas', 'seaborn', or 'matplotlib'

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Get a sample dataset
Here we'll take advantage of built-in dataset 'iris'

```python
df = sns.load_dataset('iris')
df
```

## Box Plots vs Violin Plots

### Store all variables to a list

```python
vars = df.columns.values.tolist()
vars
```

### Box Plot

```python
# make a box plot of the iris species
sns.boxplot(y=vars[0], data=df)
plt.show()
```

```python
# set a grey background color
sns.set_style("darkgrid")
# and create the box plot again to see the difference
sns.boxplot(y=vars[0], data=df)
plt.show()
```

Although visualizing all sepal lengths can be useful, grouping by species may allow for a better assessment of the data.

```python
sns.boxplot(x='species', y=df[vars[0]], data=df)
plt.ylabel(vars[0])
plt.show()
```

**Boxplots** allow for a overall perception of the distribution: Median, quartiles, or outliers will be easily identified.

Nevertheless, the underlying distribution will be hidden, as well as the number of observations in the dataset, which may, in some cases, impair our ability to draw conclusions.

To overcome this limitation, adding a stripplot may be a good  choice.

```python
# add stripplot to the box plot
sns.boxplot(x='species', y=df[vars[0]], data=df)
plt.ylabel(vars[0])
sns.stripplot(x='species', y=df[vars[0]], data=df, jitter=True, size=3, color='black') # color black allow to see the difference
plt.show()
```

An even better alternative to show the data distribution is to use a violin plot instead of a box plot

### Violin Plot

```python
#make a violin plot of all rows in a dataframe column
sns.violinplot(y=df[vars[0]])
#set up y label
plt.ylabel(vars[0])
plt.show()
```

```python
# Group variables by species
sns.violinplot(x='species', y=df[vars[0]], data=df)
plt.ylabel(vars[0])
plt.show()
```

By data visualization, we can start drawing some conclusions. The first is that the three groups seem to have different distributions of septal lengths.

**Remark**: further statistical assessment will be necessary to draw definite conclusions

```python
fig, axs = plt.subplots(1, len(vars)-1, figsize=(15,5))
for i in range(0,len(vars)-1):
    sns.violinplot(x='species', y=df[vars[i]], data=df,ax=axs[i])
    axs[i].set_ylabel(vars[i])

plt.show()
```

Finally, let's look at the differences by putting boxplots and violinplots near to each other

```python
fig, axs = plt.subplots(2, len(vars)-1, figsize=(20,10))
for i in range(0,len(vars)-1):
    sns.violinplot(x='species', y=df[vars[i]], data=df,ax=axs[0,i])
    axs[0,i].set_ylabel(vars[i])
for i in range(0,len(vars)-1):
    sns.boxplot(x='species', y=df[vars[i]], data=df,ax=axs[1,i])
    axs[1,i].set_ylabel(vars[i])
    sns.stripplot(x='species', y=df[vars[i]], data=df, jitter=True, size=3, color='black', ax=axs[1,i])

plt.show()
fig.savefig('front.png')

```
